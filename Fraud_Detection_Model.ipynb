```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
credit_data = pd.read_csv('/content/creditcard.csv')
```

```python
#v are the deatures about each transaction
# time isthe time elapsed since the first transaction
# amount in usd
# class is used to check whether the trans is legit or not (1 is fradulant)
credit_data.tail()
```

```python
#check dist of legit transac and frauds
credit_data['Class'].value_counts()
#this dataset is very unbalanced
#initial data cannot be used to predict fraud cause the fraud data points are very less
# thus pre processing is needed
```

```python
# 0 -> normal transaction
# 1 -> fraud transaction
# we seperate the data for analysis
legit = credit_data[credit_data.Class == 0]
fraud = credit_data[credit_data.Class == 1]
```

```python
fraud.shape
```

```python
# stat measure
legit.Amount.describe()
```

```python
fraud.Amount.describe()
# we can observe that mean value is much greater
```

```python
# compare values
credit_data.groupby('Class').mean()
#check the times
```

```python
# small dist ds for legit and fraud
# check dist is even (random sample)
legit_sample = legit.sample(n=492)
```

```python
df = pd.concat([legit_sample,fraud],axis = 0)
#adding row wise
```

```python
df.head()
```

```python
df.shape
```

```python
df.tail()
#fraud
```

```python
df['Class'].value_counts()
# now we have a uniform data
```

```python
df.groupby('Class').mean()
#mean is constatnt as the og so data set is still consistent i.e good sample
```

```python
#spilt daata
X = df.drop(columns = 'Class',axis = 1) #column
Y = df['Class']
```

```python
print(X)
```

```python
print(Y) #labels
```

```python
# train and test data
# 20% of data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify = Y, random_state = 2)
```

```python
print(X.shape, X_train.shape,X_test.shape)
```

```python
#model training
# log reg (binary data 1/0)
model = LogisticRegression()
```

```python
# training with data
model.fit(X_train,Y_train)
```

```python
#evaluation
# using training data
X_train_prediction = model.predict(X_train)
# compare values
train_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print('Train Accuracy :',train_accuracy)
# check score for train data
```

```python
X_test_prediction = model.predict(X_test)
test_accuracy = accuracy_score(X_test_prediction,Y_test)
# for test data
# accuracy score should somewhat match or else model is over/underfitted
```

```python
print('Test accuracy score: ',test_accuracy)
```
